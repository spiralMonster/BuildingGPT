# Imports

```python
import torch
```

```python
from gpt_model import GPTModel
```

# Model Config

```python
model_config={
    "embedding_dim":768,
    "context_length":1024,
    "vocab_size":50257,
    "n_attention_heads":12,
    "n_layers":12,
    "dropout_rate":0.1,
    "use_qkv_bias":False,
    "use_attention_dropout":False
}
```

# Model Initialization

```python
gpt_model=GPTModel(model_config)
```

```python
gpt_model
```

# Model Total Parameters

```python
total_parameters=sum(p.numel() for p in gpt_model.parameters())
print(f"[INFO] Total parameters in the model: {total_parameters}")
```

# Model Size

```python
model_size=total_parameters*4/(1024*1024)
print(f"[INFO] The size of the model is {model_size:.2f} MB.")
```
